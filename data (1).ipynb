```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load CSV file
# Replace 'sales_data.csv' with your file name
df = pd.read_csv("C:/Users/MYTHILI G/Downloads/sales_data.csv")

# Step 3: Explore data
print(df.head())
print(df.info())
print(df.describe())

# Step 4: Group by category (example: Product or Region) and sum sales
sales_by_product = df.groupby("Product")["Sales"].sum()
print(sales_by_product)

# Step 5: Plot the results
sales_by_product.plot(kind="bar", figsize=(8,5), title="Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()

# Example 2: Group by month
sales_by_month = df.groupby("Month")["Sales"].sum()
sales_by_month.plot(kind="line", marker="o", figsize=(8,5), title="Sales Trend by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```
